# Pretrained YOLO fine tunning

Check paremeters: https://docs.ultralytics.com/modes/train/#resuming-interrupted-trainings

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO

# downloads the yolo model in cwd.
model = YOLO(model='yolov8m.pt', task='detect')

model.train(
    data='data.yaml',
    epochs=200,
    patience=15,  # Epochs to wait for an improvement in the val before stopping training early.
    batch=16,
    lr0=0.01,
    imgsz=640,
    dropout=0.4,  # Default is 0.
    augment=True,  # Apply default augmentation tecnhiques.
    optimizer='auto',
    cos_lr=True,  # adjusts the learning rate for better convergence.
    pretrained=True,
    device=0,  # 0 to use first available GPU (collab).
    save=True,
    save_period=10,  # Save model checkpoints every 10 epochs (for resume).
    val=True,
    cache=True,
    plots=True,  # Default is False.
    seed=1,  # Ensures reproducibility.
    resume=False,  # Resumes training from the last saved checkpoint.
)
```
